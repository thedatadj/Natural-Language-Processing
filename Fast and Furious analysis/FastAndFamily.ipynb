## Loading and joining
I load and join the subtitles of each movie.1

```python
# Important modules
import pandas as pd
import seaborn as sns
import nltk
import matplotlib.pyplot as plt
```

```python
movies = []
movies_paths = [ff1, ff2, ff3, ff4, ff5, ff6, ff7, ff8, ff9, ff10]

for path in movies_paths:
  with open(path, "r") as file:
    try:
      subtitles = file.read()
    except:
      print(movies_paths.index(path))
    movies.append(subtitles)
```

```python
len(movies)
```

## Tokenizing

```python
dic = {}
for i in range(len(movies)):
  dic["Fast {}".format(i+1)] = nltk.word_tokenize(movies[i])
```

```python
dic.keys()
```

```python
dic["Fast 1"][:10]
```

```python
family = {}
for i in range(1,11):
  frame = pd.DataFrame(dic["Fast {}".format(i)])
  try:
    family["Fast {}".format(i)] = frame.value_counts()['family']
  except:
    family["Fast {}".format(i)] = 0
```

```python
family_data = pd.DataFrame(family.values(), index=family.keys())
```

```python
sns.barplot(data=family_data, x=family_data[0], y=family_data.index)
plt.xlabel("Frequency of the word family")
```
